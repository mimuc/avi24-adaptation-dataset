```r
library(lme4)
library(dplyr)
library(bayestestR)
library(report)
```

```r
df <- read.csv("processed/ecg_block.csv")
df <- df[df$block_number != 6.0,]
df <- df[df$block_number != 7.0,]
head(df)
```

```r
m <- lmer(ecg_rate_mean ~ block_number + (1|pid), data = df)
print(report(m), verbose=TRUE)
```

```r
m <- lmer(hrv_rmssd ~ block_number + (1|pid), data = df)
print(report(m), verbose=TRUE)
```
